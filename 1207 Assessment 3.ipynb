# Data Visualization dataframe df1 

1. top 5 neighborhoods by average price (Bar chart)

```python
average_prize=df1.groupby("neighbourhood")["price"].mean().nlargest(5)
average_prize.plot(kind="bar",color="#003f7f")
plt.title("Top 5 neighbourhood by average price")
plt.xlabel("neighbourhood")
plt.ylabel("average_price")
plt.show() 

```

Purpose: This visualization helps identify which neighborhoods have the highest average rental prices. It can be useful for understanding market trends and targeting areas for investment or marketing.

2 . Room type VS pricing (Box Plot)

```python
sns.boxplot(x="room_type",y="price",data=df1,color="red")
plt.xlabel("room_type")
plt.ylabel("price")
plt.show() 
```

Purpose: This plot displays the distribution of prices for different room types. It highlights variations in pricing and can help hosts set competitive rates based on room type and market trends.

3.Trend of listings over the year (Line chart)

```python
listings_per_year = df1["year"].value_counts().sort_index()
listings_per_year.plot(kind="line",marker="o",color="green")
plt.title("Trend if listings over the year")
plt.xlabel("year")
plt.ylabel("number of listings")
plt.show() 
```

Purpose: This line chart shows how the number of listings has changed over time. It can indicate growth in the rental market, seasonal trends, or shifts in demand, helping stakeholders make informed decisions.

4. Reviews per month by neighborhood gtoup (violin plot) 

```python
sns.violinplot(x="neighbourhood_group",y="reviews_per_month",data=df1)
plt.title("Reviews per month by neighborhood group")
plt.xlabel("neighbourhood_group")
plt.ylabel("reviews per month")
plt.show() 
```

Purpose: Compares review frequencies across neighborhood groups, providing insights into guest engagement and overall popularity.

5. reletionship between availability and price (scatter plot)

```python
sns.scatterplot(x="availability_365",y="price",data=df1)
plt.title("availability Vs price")
plt.xlabel("Availability per year")
plt.ylabel("price")
plt.show() 
```

Purpose: Analyzes the relationship between listing availability and pricing, helping to inform pricing strategies for optimal occupancy.

# Data Visualization dataframe df2

6. Salary Distribution by Department (Heat map)

```python
df2.columns
```

```python
salary_dept = df2.groupby("Department")["Salary"].mean().reset_index()
sns.heatmap(salary_dept.set_index("Department"),annot=True,cmap="viridis")
plt.title("salary distribution by department")
plt.xlabel("department")
plt.ylabel("salary")
plt.show() 
```

purpose :
This heatmap shows the average salary by department, making it easy to compare salary levels across departments. It helps identify which departments have higher or lower average salaries.

7. performance scores by marital status (count plot)

```python
ax= sns.countplot(data=df2,x="MaritalDesc",hue="PerformanceScore")
for bars in ax.containers: 
    ax.bar_label(bars) 
```

purpose :This count plot shows the distribution of performance scores by marital status. It helps identify trends in employee performance across different marital categories.

8. performance scores by gender (stacked bar chart)

```python
stacked_data = df2.groupby(['Sex', 'PerformanceScore']).size().unstack(fill_value=0)
stacked_data.plot(kind='bar', stacked=True)
plt.title('Performance Scores by Sex (Stacked Bar Chart)')
plt.xlabel('Sex')
plt.ylabel('Count of Performance Scores')
plt.show()
```

purpose: 
This stacked bar chart shows performance scores by gender, allowing easy comparison of employee counts in each score category for different genders.

9.  relationship between performance scores and employee satisfaction levels?

```python
plt.figure(figsize=(10, 6))
plt.scatter(df2['EmpSatisfaction'], df2['PerformanceScore'], s=df2['Salary'] / 1000, alpha=0.5)
plt.title('Bubble Chart of Performance Scores by Employee Satisfaction (Size by Salary)')
plt.xlabel('Employee Satisfaction')
plt.ylabel('Performance Score')
plt.show()
```

purpose :This bubble chart shows the relationship between employee satisfaction and performance scores, with bubble size indicating salary. It helps visualize trends and correlations between these variables.

10. engagement survey scores vary across different departments

```python
department_engagement = df2.groupby('Department')['EngagementSurvey'].mean()
plt.figure(figsize=(8, 8))
department_engagement.plot.pie(autopct='%1.1f%%', startangle=90, colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])
plt.title('Average Engagement Survey Scores by Department')
plt.ylabel('')  
plt.axis('equal') 
plt.show()
```


purpose : This pie chart shows average engagement survey scores by department, highlighting the proportion of engagement levels in each department for easy comparison.
